```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import plotly.express as px
import seaborn as sns
```

```python
#wczytanie dancych z pliku
df=pd.read_csv("/home/michal/AiBD/lab6/exercise.csv")
```

```python
#wybranie wierszy o znanym y
df1=df.iloc[:40]
```

```python
#wybranie wierszy o nieznanym y
df2=df.iloc[40:60]
```

```python
#stworzenie modelu
X = df1[['x1','x2']]
Y = df1['y']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
predictions = model.predict(X)
print(model.params)
predictions
```

```python
#zastąpnienie starych wartości nowymi
df1.values[:,0] = predictions
```

```python
#wyświetlenie punktów dla x1
df1.plot.scatter(x='x1',y='y',marker='o')
```

```python
#wyświetlenie punktów dla x2
df1.plot.scatter(x='x2',y='y',marker='o')
```

```python
sns.set(style="whitegrid")
#wykres residuów dla x1
sns.residplot(df1['x1'], df1['y'], lowess=True, color="r")
plt.show()
```

```python
#wykres residuów dla x2
sns.residplot(df1['x2'], df1['y'], lowess=True, color="r")
plt.show()
```

```python
#prognozy dla pozostałych 20 punktów
X1 = df2['x1']
X2 = df2['x2']
Y1 = model.params[0] + (model.params[1]) * X1 + (model.params[2]) * X2
Y1
```
