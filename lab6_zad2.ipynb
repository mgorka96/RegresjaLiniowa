```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#wczytanie danych z pliku
df=pd.read_csv("/home/michal/AiBD/lab6/beauty.csv")
```

```python
#
X = df['btystdave']
Y = df['courseevaluation']
X = sm.add_constant(X)
X
model = sm.OLS(Y, X).fit()
predictions = model.predict(X)
print(model.params)
predictions
```

```python
X1 = df['btystdave']
Y1 = model.params[0] + model.params[1] * X1
Y1
```
